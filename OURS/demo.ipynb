```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_images(image_path1, image_path2):
    """
    Displays two images side by side from the given file paths.

    Parameters:
        image_path1 (str): The path to the first image file.
        image_path2 (str): The path to the second image file.
    """
    img1 = mpimg.imread(image_path1)
    img2 = mpimg.imread(image_path2)
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(img1)
    axes[0].axis('off')  # Hide axes for better display
    axes[0].set_title('Image 1')
    
    axes[1].imshow(img2)
    axes[1].axis('off')  # Hide axes for better display
    axes[1].set_title('Image 2')
    
    plt.tight_layout()
    plt.show()

image_path1 = '/media/users/leo/workspace/UIE/datasets/UIEB_Dataset/2_img_.png'
image_path2 = '/media/users/leo/workspace/UIE/datasets/UIEB_Dataset/3_img_.png'
show_images(image_path1, image_path2)
```
